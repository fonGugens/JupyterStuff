# Фукнции, применяемые к датафреймам

```python
genres = movie[['movie_title','genres']]
```

```python
genres
```

```python
def reversator(value):
    return value[::-1]
```

Не pandas way:

```python
for row in genres.values:
    for value in row:
        print(reversator(value))
```

```python
for i, row in genres.iterrows():  # Иттерация по строкам
    print(row.map(reversator))
```

```python
for i, col in genres.iteritems():  # Иттерация по колонкам
    print(col.map(reversator))
```

# Timeseries (Временные ряды)

```python
stock = pd.read_csv('Data/amzn_stock.csv', index_col='Date', parse_dates = True)
stock.head()
```

```python
stock['2010-02':'2011-03']
```

```python
# Изменение масштаба даты, увеличить или уменьшить
stock.resample('2h').asfreq()
```

```python
stock.resample('1w').mean()
```

```python
stock.head(10)
```

Скользящее окно
Расчет нового значения происходит за счет текущего + $n$ предыдущих

```python
stock.rolling(3, min_periods=1).mean() #min_periods минимальное кол-во допустимых значений
```

Для первой - по одному значению, для второго - по двум,
Для $n$-ного значения - по $n$ строкам


```python
stock.expanding().mean()
```

```python
sns.lineplot(x='Date',y='Close', data=stock, color='yellow')
```

Exponetial axpanding window $\alpha$ - коэффициент угасания веса, чем дальше от $n$ строки
тем меньше вес.

```python
stock.ewm(alpha=0.7).mean()
```

```python
stock['Open'].plot(lw=4)
ns = stock['Open'].rolling(10,min_periods=1).mean()
ns.plot(color='yellow',lw=2)
```
